# 读取数据

## 数据处理

```python
import numpy as np
import pandas as pd
```

```python
country = pd.read_csv(r'C:\Users\Kermoci\Desktop\数据可视化\2商业图表可视化\05Hans Rosling Charts Matplotlib 绘制\Hans Rosling Charts Matplotlib 绘制\country_metadata.csv')
country = country[['Country Code', 'Region']]
country.head()
```

```python
country["Country Code"] = country["Country Code"].apply(lambda x : x.lower())
country.head()
```

```python
file01 = r"C:\Users\Kermoci\Desktop\数据可视化\2商业图表可视化\05Hans Rosling Charts Matplotlib 绘制\Hans Rosling Charts Matplotlib 绘制\GM-Life Expectancy- Dataset - v11.xlsx"

lifeExp = pd.read_excel(file01,sheet_name=3)
lifeExp.rename(columns={"geo":"Country Code"},inplace=True)
lifeExp.head()
```

```python
file02 = r"C:\Users\Kermoci\Desktop\数据可视化\2商业图表可视化\05Hans Rosling Charts Matplotlib 绘制\Hans Rosling Charts Matplotlib 绘制\GM-GDP per capita - Dataset - v26.xlsx"

GDP_per = pd.read_excel(file02,sheet_name=3)
GDP_per.rename(columns={"geo":"Country Code"},inplace=True)
GDP_per.head()
```

```python
读取 pop 数据并更改列名，为数据匹配做准备
```

```python
file03 = r"C:\Users\Kermoci\Desktop\数据可视化\2商业图表可视化\05Hans Rosling Charts Matplotlib 绘制\Hans Rosling Charts Matplotlib 绘制\Data Population - v5 - 1800 to 2100 World Regions and Countries by Gapminder.xlsx"

pop = pd.read_excel(file03,sheet_name=3)
pop.rename(columns={"geo":"Country Code"},inplace=True)
pop.head()
```

## 进行数据合并，获得绘制动态图所需数据

```python
df = pd.merge(country, lifeExp, how="left", on="Country Code")
df = pd.merge(df, GDP_per, how='left', on=['Country Code', 'time'])
df = pd.merge(df, pop, how='left', on=['Country Code', 'time'])
df
```

```python
#处理缺失值
df.dropna(axis=0, inplace=True)
df
```

```python
def region_set(x):
    if (x == 'South Asia')| (x =='East Asia & Pacific'):
        return "Asia"
    elif (x == 'Latin America & Caribbean')| (x == 'North America'):
        return "The Americas"
    elif (x == 'Sub-Saharan Africa')| (x =='Middle East & North Africa'):
        return "Africa"
    elif x == 'Europe & Central Asia':
        return "Europe"
    
def color_set02(x):
    if (x == 'South Asia')| (x =='East Asia & Pacific'):
        return "#FF5872"
    elif (x == 'Latin America & Caribbean')| (x == 'North America'):
        return "#7FEB00"
    elif (x == 'Sub-Saharan Africa')| (x =='Middle East & North Africa'):
        return "#00D5E9"
    elif x == 'Europe & Central Asia':
        return "#FFE700"
```

```python
select = ["Country Code","Region","name","time","Life expectancy","Income per person",
          "GDP total", "population"]

df = df[select]
df['region'] = df['Region'].apply(lambda x : region_set(x))
df['color'] = df['Region'].apply(lambda x : color_set02(x))
del df['Region']
df
```

```python
df.to_csv(r"C:\Users\Kermoci\Desktop\数据可视化\2商业图表可视化\05Hans Rosling Charts Matplotlib 绘制\Hans Rosling Charts Matplotlib 绘制\gapminderDataall.csv",
                index=None)
```

# 数据可视化

## 导入数据

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.animation as animation

gapminder = pd.read_csv(r"C:\Users\Kermoci\Desktop\数据可视化\2商业图表可视化\05Hans Rosling Charts Matplotlib 绘制\gapminderDataall.csv")
gapminder = gapminder.loc[gapminder['time']<=2019,:]
#数据整数化(int)处理，为后面动图选取数据做准备
gapminder['time'] = gapminder.time.apply(lambda x : int(x))
gapminder.head()
```

## 开始绘图

```python
fig, ax = plt.subplots(figsize = (12,7),dpi=150)
data_test = gapminder[gapminder.time == 1800]
col = [i for i in data_test['color']]
#定义国家标注：以美国和中国为主
china = data_test.loc[data_test.name == 'China',['Life expectancy','Income per person']]
chinaloc = china.values.tolist()
usa = data_test.loc[data_test.name == 'United States',['Life expectancy','Income per person']]
usaloc = usa.values.tolist()
#美化图表
#设置轴脊(spines)显示情况
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#设置刻度显示范围
ax.set_ylim(ymin = 0,ymax = 90)
ax.set_xlim(xmin = 100,xmax = 10**5.5)
#ax.set_xticks(np.arange(1800,2050,0))
#设置x轴刻度比例
ax.set_xscale("log")

ax.tick_params(axis='both', colors='gray',labelsize = 15)
ax.grid(axis = 'both',lw = 1,alpha = .8,color = "#E3EEF0")
#更改刻度是否显示
ax.tick_params(bottom = False, left = False)
scatter = ax.scatter(data_test["Income per person"],data_test["Life expectancy"],
           s = data_test["population"]/350000,c = col,
           edgecolors = "black",linewidths = 1)
#设置label
ax.set_xlabel('Incom', color='#425663',fontsize = 19)
ax.set_ylabel('Life expectancy', color='#425663',fontsize = 19)
#ax.set_text()
#设置文本位置相对固定：text(transform = ax.transAxes)
ax.text(.5, .5, 1800, va='center', ha='center', alpha=.1,size=200,transform = ax.transAxes,
        fontfamily = "Franklin Gothic Book")
ax.text(.7,0.02,"per person(GDP/captita,PPP$ inflation-adjusted)",size=12,
        transform = ax.transAxes,color = "gray",ha='center')
ax.text(0.01,.9,"years",size=13,color = "gray",rotation=90,transform = ax.transAxes)
ax.text(.3, .95, "INCOME LEVEL 1",color = "gray",alpha=.8,size = 15,ha="center",
       transform = ax.transAxes)
ax.text(.6, .95, "LEVEL 2",color = "gray",alpha=.8,size = 15,ha="center",
       transform = ax.transAxes)
ax.text(.8, .95, "LEVEL 3",color = "gray",alpha=.8,size = 15,ha="center",
       transform = ax.transAxes)
#添加解释文本
ax.text(.96, .38, "Size: Population",color = "gray",alpha=.8,size = 17,ha="center",
       transform = ax.transAxes, rotation = 90)

#设置网格等属性置于底层
ax.set_axisbelow(True)

#添加图例：新方法
data_test_group = data_test.groupby("region")
#定义颜色(重要步骤)
colors = {
        'Africa': '#00D5E9',
        'Asia': '#FF5872',
        'Europe': '#FFE700',
        'The Americas': '#7FEB00',
    }
region_name = ['Africa','Asia','Europe','The Americas']
for name in region_name:
    ax.scatter([],[],
               c = colors[name],
               label= name,
               edgecolors = "black",
               linewidths = 1)
#添加图例结束
#添加标题
ax.annotate("China", xy = (chinaloc[0][1],chinaloc[0][0]),c = "#E5F100",
            xycoords="data",xytext=(-50,5), textcoords='offset points',
            ha='center',bbox = dict(boxstyle="round",fc = "red",ec="black"),
            arrowprops=dict(arrowstyle="->", ec="black"))

ax.annotate("USA", xy = (usaloc[0][1],usaloc[0][0]),
            xycoords="data",c="white",
            xytext=(-50,5), textcoords='offset points',
            ha='center',bbox = dict(boxstyle="round",fc = "#002664",ec="black"),
            arrowprops=dict(arrowstyle="->",ec="#002664"))
#对图例进行详细设置
scatter_legend = ax.legend(fontsize=12,framealpha = .5,markerscale =2,title = 'Region',
                          loc = 'lower right')
scatter_legend.get_title().set_fontsize(fontsize = 15)#面向对象
```

## 绘制动图

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
#自定义刻度设置
from matplotlib.ticker import MultipleLocator, FuncFormatter
import matplotlib.ticker as ticker
plt.rcParams['font.family'] = ['Roboto Mono']
#设置动态上限
plt.rcParams['animation.embed_limit'] = 2**128
fig, ax = plt.subplots(figsize = (12,7),dpi=150)

def animation_bubble(current_year):
    data_test = gapminder[gapminder.time == current_year]
    col = [i for i in data_test['color']]
    #定义国家标注：以美国和中国为主
    china = data_test.loc[data_test.name == 'China',['Life expectancy','Income per person']]
    chinaloc = china.values.tolist()
    usa = data_test.loc[data_test.name == 'United States',['Life expectancy',
                                                           'Income per person']]
    usaloc = usa.values.tolist()
    ax.clear()
    #美化图表
    #设置轴脊(spines)显示情况
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    #设置刻度显示范围
    ax.set_ylim(ymin = 0,ymax = 90)
    ax.set_xlim(xmin = 100,xmax = 10**5.5)
    #ax.set_xticks(np.arange(1800,2050,0))
    #设置x轴刻度比例
    ax.set_xscale("log")
    ax.tick_params(axis='both', colors='gray',labelsize = 15)
    ax.grid(axis = 'both',lw = 1,alpha = .8,color = "#E3EEF0")
    #更改刻度是否显示
    ax.tick_params(bottom = False, left = False)
    scatter = ax.scatter(data_test["Income per person"],data_test["Life expectancy"],
               s = data_test["population"]/350000,c = col,
               edgecolors = "black",linewidths = 1)
    #设置label
    ax.set_xlabel('Incom', color='#425663',fontsize = 19)
    ax.set_ylabel('Life expectancy', color='#425663',fontsize = 19)
    #设置文本位置相对固定：text(transform = ax.transAxes)
    ax.text(.5, .5, current_year, va='center', ha='center', alpha=.1,size=200,
            transform = ax.transAxes,fontfamily = "Franklin Gothic Book")
    ax.text(.7,0.02,"per person(GDP/captita,PPP$ inflation-adjusted)",size=12,
            transform = ax.transAxes,color = "gray",ha='center')
    ax.text(0.01,.9,"years",size=13,color = "gray",rotation=90,transform = ax.transAxes)
    ax.text(.3, .95, "INCOME LEVEL 1",color = "gray",alpha=.8,size = 15,ha="center",
           transform = ax.transAxes)
    ax.text(.6, .95, "LEVEL 2",color = "gray",alpha=.8,size = 15,ha="center",
           transform = ax.transAxes)
    ax.text(.8, .95, "LEVEL 3",color = "gray",alpha=.8,size = 15,ha="center",
           transform = ax.transAxes)
    #添加解释文本
    ax.text(.96, .38, "Size: Population",color = "gray",alpha=.8,size = 17,ha="center",
       transform = ax.transAxes, rotation = 90)
    #设置网格等属性置于底层
    ax.set_axisbelow(True)
    #添加图例：
    data_test_group = data_test.groupby("region")
    #定义颜色(重要步骤)
    colors = {
            'Africa': '#00D5E9',
            'Asia': '#FF5872',
            'Europe': '#FFE700',
            'The Americas': '#7FEB00',
        }
    region_name = ['Africa','Asia','Europe','The Americas']
    for name in region_name:
        ax.scatter([],[],
                   c = colors[name],
                   label= name,
                   edgecolors = "black",
                   linewidths = 1)
    #添加图例结束
    ax.annotate("China", xy = (chinaloc[0][1],chinaloc[0][0]),c = "#E5F100",
                xycoords="data",xytext=(-50,5), textcoords='offset points',
                ha='center',bbox = dict(boxstyle="round",fc = "red",ec="black"),
                arrowprops=dict(arrowstyle="->", ec="black"))

    ax.annotate("USA", xy = (usaloc[0][1],usaloc[0][0]),
                xycoords="data",c="white",
                xytext=(-50,5), textcoords='offset points',
                ha='center',bbox = dict(boxstyle="round",fc = "#002664",ec="black"),
                arrowprops=dict(arrowstyle="->",ec="#002664"))
    #对图例进行详细设置
    scatter_legend = ax.legend(fontsize=12,framealpha = .5,markerscale =2,
                               title = 'Region',
                              loc = 'lower right')
    scatter_legend.get_title().set_fontsize(fontsize = 15)#面向对象

#animation_bubble(2019)
bubble_animator = animation.FuncAnimation(fig, animation_bubble,frames=np.arange(1800,2020), 
                                          interval = 200)
# #animator.save("gapminder")
HTML(bubble_animator.to_jshtml())
#animator.save('gapminder.gif',dpi=300, writer='imagemagick')
```
