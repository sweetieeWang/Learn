```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

iris = load_iris()
iris = pd.DataFrame(iris.data, columns=['SpealLength', 'SpealWidth', 'PetalLength', 'PetalWidth'])
iris.head()

iris.head()
```

```python
#1.散点图
#计算面积，用面积来代表大小
sepal_area = []
petal_area = []
for i in range(len(iris)):
    sepal_area.append(iris.SpealLength[i] * iris.SpealWidth[i])
    petal_area.append(iris.PetalLength[i] * iris.PetalWidth[i])
fig,ax = plt.subplots()
ax.set_title("Area of the sepal and petal") 
ax.set_xlabel("sepal_area") 
ax.set_ylabel("petal_area") 
ax.scatter(sepal_area,petal_area,alpha=0.5, c = '#9467bd') 
plt.grid(True)
plt.show()
```

```python
#2.箱图/小提琴图
fig, axes = plt.subplots(figsize=(12, 5))
axes.violinplot(iris, 
                   showmeans=False,
                   showmedians=True
                   )
axes.set_title('violin plot')

# adding horizontal grid lines

axes.yaxis.grid(True)
axes.set_xlabel('xlabel')
axes.set_ylabel('ylabel')

plt.setp(axes,xticklabels=iris.columns)

plt.show()
```

```python
# 3.六角箱图或者核密度估计
import seaborn as sns
sns.jointplot(x = sepal_area, y = petal_area, kind="hex")
sns.jointplot(x = sepal_area, y = petal_area, kind="kde") 
```
